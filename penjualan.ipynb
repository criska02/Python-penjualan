```python
import mysql.connector
import pandas as pd
class Barang:
    
    def __init__(self, id_barang, nama_barang, kuantitas, harga_beli, harga_jual):        
        self.id_barang = id_barang
        self.nama_barang = nama_barang
        self.kuantitas = kuantitas
        self.harga_beli = harga_beli
        self.harga_jual = harga_jual
        
   
    def tampilkan_data_barang(self): 
        print("============Data Barang============")
        print("ID Barang:", self.id_barang)
        print("Nama Barang:", self.nama_barang)
        print("Kuantitas:", self.kuantitas)
        print("Harga Beli:", self.harga_beli)
        print("Harga Jual:", self.harga_jual)

class Pegawai:
    def __init__(self, id_pegawai, nama_pegawai, no_hp):        
        self.id_pegawai = id_pegawai
        self.nama_pegawai = nama_pegawai
        self.no_hp = no_hp
    
    def tampilkan_data_pegawai(self): 
        print("============Data Pegawai============")
        print("ID Pegawai :", self.id_pegawai)
        print("Nama Pegawai :", self.nama_pegawai)
        print("No HP :", self.no_hp)        
        
class Penjualan(Barang, Pegawai):
    def __init__(self, id_penjualan, barang, pegawai, tanggal_penjualan, kuantitas_jual):
        Barang.__init__(self, barang.id_barang, barang.nama_barang, barang.kuantitas, barang.harga_beli, barang.harga_jual)
        Pegawai.__init__(self, pegawai.id_pegawai, pegawai.nama_pegawai, pegawai.no_hp)
        
        self.id_penjualan = id_penjualan
        self.tanggal_penjualan = tanggal_penjualan
        self.kuantitas_jual = kuantitas_jual  
        self.harga_jual = barang.harga_jual
        self.jumlah = kuantitas_jual*barang.harga_jual
        
        print("==========Penjualan==========")
        print("waktu : ", self.tanggal_penjualan)
        print("Kuantitas jual : ", self.kuantitas_jual)
        print("Harga jual : ", self.harga_jual)
        print("Jumlah : ", self.jumlah)

        super().tampilkan_data_barang()
        super().tampilkan_data_pegawai()
        
class Pembelian(Barang, Pegawai):
    def __init__(self, id_pembelian, barang, pegawai, tanggal_pembelian):       
        Barang.__init__(self, barang.id_barang, barang.nama_barang, barang.kuantitas, barang.harga_beli, barang.harga_jual)
        Pegawai.__init__(self, pegawai.id_pegawai, pegawai.nama_pegawai, pegawai.no_hp)
        
        self.id_pembelian = id_pembelian
        self.tanggal_pembelian = tanggal_pembelian
        self.kuantitas_beli = barang.kuantitas
        self.harga_beli = barang.harga_beli
        self.jumlah = self.kuantitas_beli*barang.harga_beli

        
        print("==========Pembelian==========")
        print("Tanggal Pembelian : ",self.tanggal_pembelian)
        print("Kuantitas Beli : ",self.kuantitas_beli)
        
        super().tampilkan_data_barang()
        super().tampilkan_data_pegawai()
            
```
